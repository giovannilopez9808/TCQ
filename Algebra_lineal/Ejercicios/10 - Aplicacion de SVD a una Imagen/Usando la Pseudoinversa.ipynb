```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
np.set_printoptions(suppress=True) # pedimos a numpy que no muestre los numeros pequenos de punto flotante
```

```python
# En la clase de sistemas de ecuaciones lineales, vimos que un sistema podia tener 1 solucion, 0 o infinitas
# Cuando tenemos un sistema sobredeterminado, o sea mas ecuaciones que incognitas, no existe solucion
# La pseudoinversa puede ayudarnos a encontrar una solucion que minimice el error en el sentido de minimo error
# cuadratico
```

```python
# Veamos un ejemplo
x = np.linspace(-5, 5, 1000)
y_1 = -2*x + 2
y_2 =  4*x + 8
y_3 = -1*x + 2

plt.plot(x, y_1)
plt.plot(x, y_2)
plt.plot(x, y_3)
plt.xlim(-2., 1)
plt.ylim(1, 5)
plt.show()
```

```python
# Claramente el sistema no tiene solucion
# Pasemos a la forma matricial del sistema de ecuaciones
matriz = np.array([[-2, -1], [4, -1], [-1, -1]])
print(matriz)
```

```python
# Calculemos una pseudoinversa
matriz_pse = np.linalg.pinv(matriz)
print(matriz_pse)

# Recordemos que cuando usamos la funcion np.linalg.pinv estamos usando por detras la descomposicion SVD
```

```python
b = np.array([[-2], [-8], [-2]])
print(b)
```

```python
resultado = matriz_pse.dot(b)
print(resultado)
```

```python
plt.plot(x, y_1)
plt.plot(x, y_2)
plt.plot(x, y_3)
plt.xlim(-2., 1)
plt.ylim(1, 5)

plt.scatter(resultado[0], resultado[1])

plt.show()
```

```python
# Tal vez sea contraintuitivo porque uno esperaria que la solucion estuviera en el centro, pero debido a que las
# ecuaciones no tienen la misma escala eso no ocurre.
```
