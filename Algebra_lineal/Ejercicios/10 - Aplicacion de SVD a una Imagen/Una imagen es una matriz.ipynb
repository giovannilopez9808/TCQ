```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# Ya vimos que una imagen para nuestro computador es una matriz, o multiples matrices, una para cada canal
# La descomposicion en valores singulares tiene aplicaciones para la compresion de imagenes, y esto es super
# importante para cuando debemos transmitir las imagenes o incluso almacenarlas.
```

```python
# Una imagen de la fotografa Frida Bredesen de una Cebra
plt.style.use('classic')
imagen = Image.open(".\\imagenes\\imagen_ejemplo_frida_bredesen.jpg")
plt.imshow(imagen)
```

```python
# Vamos a convertir la imagen a una escala de grises
imagen_gr = imagen.convert('LA')
print(imagen_gr)
```

```python
# Ahora podemos convertir la imagen en un array de numpy
imagen_mat = np.array(list(imagen_gr.getdata(band=0)), float)
print(imagen_mat) # es una array, este tipo de transformaciones muchas veces se hace para poder alimentar a una
# red neuronal, o incluso a is hacemos deeplearning, pero siempre debemos volver a la forma original para que tenga
# significado para nosotros y no para una maquina.
```

```python
# Le damos nuevamente la forma original
imagen_mat.shape = (imagen_gr.size[1], imagen_gr.size[0])
print(imagen_mat)
# Como solamente estamos con grises nos queda una unica matriz
print(imagen_mat.shape)
```

```python
plt.imshow(imagen_mat, cmap='gray')
plt.show()
```

```python
# Vemos que la imagen  no sufrio modificaciones aun con la transformacion que realizamos

imagen_mat_2 = imagen_mat / 10
print(imagen_mat_2)
#Dividimos todos los valores por 10
```

```python
plt.imshow(imagen_mat_2)
plt.show()
```
